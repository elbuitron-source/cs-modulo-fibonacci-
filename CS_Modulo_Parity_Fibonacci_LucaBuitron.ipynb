## Exercise 1 — Modulus Operator Demo

The modulus operator (%) returns the remainder after dividing one number by another.
It is useful to check things like whether a number is even or odd.


```python
print(10 % 2)    # remainder is 0
print(10 % 3)    # remainder is not 0
print(25 % 5)    # remainder is 0
print(25 % 6)    # remainder is not 0
print(157 % 12)  # larger number example
print(9 % 4)     # extra example
```

```python
n = int(input("Enter a number: "))
m = int(input("Enter another number: "))

remainder = n % m

print("The remainder of", n, "divided by", m, "is", remainder)
```

## Exercise 3 — Product Parity (Even or Odd)

```python
val1 = int(input("Enter first value: "))
val2 = int(input("Enter second value: "))

product = val1 * val2
print("The product is:", product)

if product % 2 == 0:
    print("The product is even.")
else:
    print("The product is odd.")
```

## Exercise 4 — Fibonacci Series Generator

```python
terms = int(input("How many Fibonacci terms do you want? "))

a, b = 1, 1

for i in range(terms):
    print(a)
    a, b = b, a + b
```
